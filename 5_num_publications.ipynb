```python
df = pd.DataFrame()

# create date range from 2014 to 2023 and put in columns
df['year'] = pd.date_range(start='2012', end='2023', freq='Y').year
# piut # of publications in another column
df['twitter_publications'] = [1837, 2472, 3137, 3638, 4099, 4923, 5528, 5731, 7623, 9052, 9045]
df['twitter_publications_l'] = [1396, 1812, 2183, 2476, 2785, 3277, 3608, 3626, 4817, 5603, 5624]
df['youtube_publications'] = [705, 908, 1026, 1040, 1216, 1500, 1981, 2548, 4208, 5581, 5990]
df['youtube_publications_l'] = [433, 576, 610, 580, 678, 852, 1068, 1269, 1789, 2392, 2561]

# df['reddit_publications'] = [740, 952, 962, 1296, 1643, 2821, 4301, 4560, 6392, 7885, 8902]
```

```python
# plot the data with seaborn
fig, ax = plt.subplots(figsize=(10, 5))
# change theme to lighter grid
sns.set_theme(style="ticks")
sns.lineplot(data=df, x='year', y='twitter_publications_l', ax=ax, label='Twitter', color='skyblue')
# sns.lineplot(data=df, x='year', y='reddit_publications', ax=ax, label='Reddit')
sns.lineplot(data=df, x='year', y='youtube_publications_l', ax=ax, label='Youtube', color='red')
ax.set_xlabel('Year')
ax.set_ylabel('# of Publications')
plt.legend()
plt.show()
```
